```python
# List and Tuples
tuples1=("disco",10,1.2,"hard rock",10) #Tuple can be stored int,float,string. Tuples are immutable
print(tuples1[:3])
Ratings=(10,9,6,5,10,8,9,6,2)
RatingsSorted=sorted(Ratings)
print(RatingsSorted)
Ratings1=Ratings
print(Ratings1)
tuples2=(1,2,("pop","rock"),(3,4),("disco",(1,2)))
print(tuples2[2][1])
print(tuples2[2][1][1])

# Lists are mutable
L=["Michael Jackson",10.1,1982]
L.extend(["pop",10])
print(L)
#Append function can be added one member of the list.
L.append(["Religious",15])
print(L)

#Lists are mutable,it means it can be added values
A=["disco",10,1.2]
print(A)
A[0]="Hard Rock"
print(A)
del(A[1])
print(A)

#We can convert list into string using split() function
B="hard rock"
print(type(B))
B.split(",")
print(B)
print("A,B,C,D".split(","))
```
